# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to explore, inspect, and visualize the Iris dataset using Python libraries such as Pandas, Matplotlib, and Seaborn.

## Dataset
Iris Dataset

## Skills Demonstrated
- Data loading and preprocessing
- Exploratory Data Analysis (EDA)
- Data visualization
- Statistical analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset directly from seaborn
iris = sns.load_dataset("iris")
```

```python
# Print shape of dataset
print("Dataset Shape:")
print(iris.shape)

print("\nColumn Names:")
print(iris.columns)
```

```python
# Display first 5 rows
print("\nFirst 5 Rows:")
print(iris.head())
```

```python
print("\nDataset Information:")
print(iris.info())

print("\nSummary Statistics:")
print(iris.describe())
```

```python
# Scatter plot showing relationship between sepal length and sepal width
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette="deep"
)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.show()
```

```python
# Histograms for all numerical features
iris.hist(figsize=(10,8), bins=15)

plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
#Box plots to identify outliers
plt.figure(figsize=(12,6))

iris_melted = iris.melt(id_vars="species")
# Convert dataset to long format for easier plotting

sns.boxplot(
    data=iris_melted,
    x="variable",
    y="value"
)

plt.title("Box Plot of Iris Features")
plt.xlabel("Features")
plt.ylabel("Values")

plt.show()
```
