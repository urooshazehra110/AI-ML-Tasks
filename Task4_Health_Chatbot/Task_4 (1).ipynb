```python
from transformers import pipeline
```

```python
chatbot = pipeline(
    "text-generation",
    model="mistralai/Mistral-7B-Instruct-v0.1",
    device_map="auto"
)

print("Model Loaded Successfully!")
```

```python
def safety_filter(user_input):

    dangerous_keywords = [
        "suicide",
        "kill",
        "overdose",
        "self-harm",
        "harm myself",
        "poison"
    ]

    user_input_lower = user_input.lower()

    for word in dangerous_keywords:
        if word in user_input_lower:
            return False

    return True
```

```python
def health_chatbot(user_query):

    # Safety Check
    if not safety_filter(user_query):
        return (
            "I'm sorry, but I cannot assist with harmful or dangerous medical requests. "
            "Please contact a healthcare professional."
        )

    # Prompt Engineering
    prompt = f"""
    Act like a helpful and friendly medical assistant.

    Guidelines:
    - Answer clearly and simply.
    - Give only general health information.
    - Do not provide dangerous medical advice.
    - Encourage users to consult a doctor for serious issues.

    User Question:
    {user_query}

    Response:
    """

    response = chatbot(
        prompt,
        max_new_tokens=150,
        temperature=0.7,
        do_sample=True,
        pad_token_id=chatbot.tokenizer.eos_token_id

    )

    generated_text = response[0]['generated_text']

    # Extract only chatbot response
    answer = generated_text.split("Response:")[-1].strip()

    return answer
```

```python
question = "What causes a sore throat?"

answer = health_chatbot(question)

print("User:", question)
print("\nChatbot:", answer)
```

```python
print("Health Chatbot")
print("Type 'exit' to stop.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Stay healthy and take care!")
        break

    response = health_chatbot(user_input)

    print("\nChatbot:", response)
    print()
```
